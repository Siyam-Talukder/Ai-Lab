```python
def bonus(li):
   new_list =[]
    for value in li:
       new_list.append( value + 0.2)
    return new_list

cgpa =[3.97,3.94,3.00,3.05,3.50]
new_cgpa=bonus(cgpa)
print(new_cgpa)
```

```python
def bonus(li):
    new_list =[]
    for i in range(len(li)):
        new_list.append( li[i] + 0.02)
    return new_list
cgpa =[3.97,3.94,3.00,3.05,3.50]
new_cgpa= bonus(cgpa)
print(new_cgpa)
```

```python
def sum(li1, li2):
    new=[]
    for i in range(len(li1)):
      new.append(li1[i]+li2[i])
    return new

li1=[10,20,30,40]
li2=[20,30,40,50]
ans=sum(li1,li2)
print(ans)
```

```python
import math
def magnitude(p):
    c=0
    for i in range (len(p)):
        c=c+p[i]**2
    x=math.sqrt(c)
    print (x)

def distance(p,q):
    c=0
    for i in range (len(p)):
        c=c+(p[i]-q[i])**2
    x=math.sqrt(c)
    print (x)

p1=[3,4]
p2=[4,5]
magnitude(p1)
distance(p1,p2)
```
