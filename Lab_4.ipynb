# Create identity matrix

```python

def createIdentityMatrix(row,col):
    m=[]
    for i in range(row):
        r=[]
        for j in range(col):
            if(i==j):
               r.append(1)
            else :
               r.append(0)
        m.append(r)
    return m

def printMatrix(m):
    for r in m:
        print(r)

g=createIdentityMatrix(5,5)
printMatrix(g)
   
```
