```python
f = open("graph.txt","w")
f.write("6 9"+"\n")
f.write("0 1"+"\n")
f.write("0 2"+"\n")
f.write("1 2"+"\n")
f.write("1 3"+"\n")
f.write("2 3"+"\n")
f.write("3 4"+"\n")
f.write("4 0"+"\n")
f.write("4 1"+"\n")
f.write("4 5"+"\n")
f.close()
```

```python
f = open("graph.txt","r")
vertex_edges = f.readline()
n_m = vertex_edges.split()
n = int(n_m[0])
m = int(n_m[1])
print(n,m)
f.close()
```

```python
def createMatrix(row,col):
    m=[]
    for i in range(row):
        r=[]
        for j in range(col):
            r.append(0)
        m.append(r)
    return m

def printMatrix(m):
    for r in m:
        print(r)

f = open("graph.txt","r")
vertex_edges = f.readline()
n_m = vertex_edges.split()
n = int(n_m[0])
m = int(n_m[1])
print(n,m)

G = createMatrix(n,n)

for i in range(m):
    line = f.readline()
    u_v = line.split()
    u = int(u_v[0])
    v = int(u_v[1])
    G[u][v] = 1

printMatrix(G)
f.close()
```

```python
f = open("undirected_graph.txt","w")
f.write("5 5"+"\n")
f.write("0 1"+"\n")
f.write("0 2"+"\n")
f.write("1 2"+"\n")
f.write("2 3"+"\n")
f.write("2 4"+"\n")
f.close()
```

```python
def createMatrix(row,col):
    m=[]
    for i in range(row):
        r=[]
        for j in range(col):
            r.append(0)
        m.append(r)
    return m

def printMatrix(m):
    for r in m:
        print(r)

f = open("undirected_graph.txt","r")
vertex_edges = f.readline()
n_m = vertex_edges.split()
n = int(n_m[0])
m = int(n_m[1])
print(n,m)

G = createMatrix(n,n)
for i in range(m):
    line = f.readline()
    u_v = line.split()
    u = int(u_v[0])
    v = int(u_v[1])
    G[u][v] = 1
    G[v][u] = 1

printMatrix(G)
f.close()
```
